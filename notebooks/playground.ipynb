```python
import importlib
import tiktoken
IdTokenBiMap = importlib.import_module("utf_token").IdTokenBiMap
tokenizer = tiktoken.get_encoding("o200k_base")
# fromuuid = importlib.import_module("utf_token").fromuuid # Forward-only method import
```

```python
test_codec = IdTokenBiMap(vocab="o200k")
hex_str = "215aada34d0987ebfb9de132d913e46b"
token_hex = test_codec.fromhex(hex_str)
print(token_hex)
```

```python
token_ids_hex = tokenizer.encode(hex_str)
token_ids_utftoken = tokenizer.encode(token_hex)
print("Hex tokens: ", len(token_ids_hex))
print("utf-tokens: ", len(token_ids_utftoken))
```

```python
print([tokenizer.decode([id]) for id in token_ids_hex])
print([tokenizer.decode([id]) for id in token_ids_utftoken])
```

```python
# Self healing test
test_codec.tohex(token_hex + "%") == hex_str
# Containment check test
if token_hex in test_codec:
    assert hex_str == test_codec.tohex(token_hex), "This will not print"
if (token_hex + "%") in test_codec:
    print("This will not print")
```
